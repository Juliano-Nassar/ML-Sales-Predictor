# Análise dos dados

```python
#Importando os dados
pasta = 'Dados-originais/'

calendar = pd.read_csv(pasta+'calendar.csv')
sales_train_evaluation = pd.read_csv(pasta+'sales_train_evaluation.csv')
sales_train_validation = pd.read_csv(pasta+'sales_train_validation.csv')
sample_submission = pd.read_csv(pasta+'sample_submission.csv')
sell_prices = pd.read_csv(pasta+'sell_prices.csv')
```

```python
def print_k_s(db,name):
    rows,columns = db.shape
    
    print("""
    ------------- {} -------------------
    Keys: 
    {}
    ----------------------------------------------
    rows = {}
    columns = {}
    
    ----------------------------------------------
    """.format(name,db.keys(),rows,columns))
```

```python
print_k_s(calendar,'Calendar')
```

```python
print_k_s(sales_train_evaluation,'sales_train_evaluation')
```

```python
print_k_s(sales_train_validation,'sales_train_validation')
```

```python
print_k_s(sample_submission,'sample_submission')
```

```python
print_k_s(sell_prices,'sell_prices')
```
