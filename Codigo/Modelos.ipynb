```python
def LSTM_pre_processing(X,y,days):
    past_info = []
    labels = []
    for i in range(days):
        day = i+1
        for feature in X.columns:
            labels.append(f"{feature} (t - {day})")
            past_info.append(X[feature].shift(day))
            
    return past_info
        
```

```python
 teste = LSTM_pre_processing(X,y,2)
```

```python
X_val, y_val, X_train, y_train = X_full_train[-28:], y_full_train[-28:], X_full_train[:-28], y_full_train[:-28]
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.models.Sequential([
    keras.Input(shape=(27)),
    keras.layers.Dense(500, activation="relu"),
    keras.layers.Dense(500, activation="relu"),
    keras.layers.Dense(200, activation="relu"),
    keras.layers.Dense(30, activation="relu"),
    keras.layers.Dense(1, activation="exponential")
])
```

```python
model.layers
```

```python
model.summary()
```

```python
model.compile(loss="mean_squared_error",
              optimizer="adam",
              metrics=["mean_squared_error"])
```

```python
history = model.fit(X_train, y_train, epochs=100,
                    validation_data=(X_val, y_val))
```
