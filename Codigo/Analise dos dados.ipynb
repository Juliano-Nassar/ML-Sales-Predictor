```python
def print_k_s(db,name):
    rows,columns = db.shape
    
    print("""
    ------------- {} -------------------
    Keys: 
    {}
    ----------------------------------------------
    rows = {}
    columns = {}
    
    ----------------------------------------------
    """.format(name,db.keys(),rows,columns))
```
